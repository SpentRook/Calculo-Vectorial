```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d


fig = plt.figure()
ax = plt.axes(projection='3d')

x, y, z = np.meshgrid(np.linspace(-1, 2, 10), 
                      np.linspace(-1, 2, 10), 
                      np.linspace(0, 1, 10)) 

u = y*z*np.exp(x*z)  #Primera componente del campo vectorial
v = np.exp(x*z) #Segunda componente del campo vectorial
w = x*y*np.exp(x*z) #Tercera componente del campo vectorial

t = np.linspace(0,1,15) 

x_1 = np.sqrt(t)
y_1 = t+1
z_1 = t**2

plt.plot(x_1, y_1, z_1)

ax.quiver(x, y, z, u, v, w, length=0.01, color = 'green') #Genera el campo vectorial en en espacio

plt.show() #Grafica el campo vectorial
```
